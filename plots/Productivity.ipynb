```python
# Create some mock data

df = pd.read_csv('./data/productivity.csv')
N = len(df['year'])
ind = np.arange(N)

data_transistor   = df['transistor'].values.tolist()
data_productivity = df['productivity'].values.tolist()


# fig, ax1 = plt.subplots()
fig, ax1 = plt.subplots(figsize=(15, 5))


plt.xlim(0, 6)
plt.ylim(0, 6.9)

ax1.set_xticklabels(df['year'])


ax1.plot(data_transistor, color='tab:red', linewidth=3)

plt.grid(linestyle='-', which='major', color='darkgray')
plt.gca().xaxis.grid(False)


ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
ax1.set_yticklabels(['','10', '100', '1000', '10,000', '100,000', '1,000,000'])

ax2 = ax1.twinx()
ax2.plot(data_productivity, color='tab:blue', linestyle='--', linewidth=3)
ax2.set_ylim([0,6.9])
ax2.set_yticklabels(['','','1', '10', '100', '1000', '10,000'])

ax1.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    direction='in',
    color='darkgray')

ax2.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    direction='in',
    color='darkgray')

ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)

ax1.tick_params(axis='both', which='major', labelsize=16)
ax2.tick_params(axis='y', which='major', labelsize=16)

ax1.text(2.4, 5.4, "Design complexity", fontsize=20, rotation=0)
ax2.text(3.4, 1.34, "Designer productivity", fontsize=20, rotation=0)

ax1.text(-1.1, 7, "Logic transistors\nper chip\n(thousands)", fontsize=20, rotation=0)
ax1.text(6.1, 7, "Designer productivity\ntransistors per\nperson per month", fontsize=20, rotation=0)


fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
fig.savefig('productivity.pdf',
            facecolor=fig.get_facecolor(), bbox_inches='tight')
```
